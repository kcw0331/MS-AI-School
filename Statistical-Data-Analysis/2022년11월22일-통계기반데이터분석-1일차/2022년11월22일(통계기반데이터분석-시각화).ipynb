## 1.2 바 그래프

```python
data = {
    '바나나' : 3, '사과' : 2, '오랜지' : 10
}
# 선 : plot, 바 : bar
plt.bar([0, 1, 2], data.values())
print(data.keys(), data.values())
plt.title('과일 개수')
plt.xticks([0, 1, 2], data.keys())
plt.yticks(sorted(data.values())) # sorted는 정렬을 해주는 것이다.
plt.grid(True)
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

### 세로바 그래프

```python
data = {
    '바나나' : 3, '사과' : 2, '오랜지' : 10
}
# 선 : plot, 바 : bar, 세로바 : barh
# print(data.keys(), data.values())
# x 값 y 값 변경

plt.barh([0, 1, 2], data.values())
plt.title('과일 개수')
plt.yticks([0, 1, 2], data.keys())
plt.xticks(sorted(data.values())) # sorted는 정렬을 해주는 것이다.
plt.grid(True)
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

## 1.3 파이차트

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Graphs']
explode = [0.05, 0.1, 0.05, 0.05]
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']
# 파이차트 메서드 plt.pie
plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle = 260, counterclock = False, shadow = True, explode = explode ,colors = colors) # 시계 방향으로 갈건지 시계 반대 방향으로 갈건지에 대한 것이다.
plt.show()
```

```python
wedgeprops = {'width' : 0.7, 'edgecolor': 'w', 'linewidth' : 5}
plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle = 260, 
        counterclock = False, shadow = True, explode = explode ,colors = colors, wedgeprops = wedgeprops) # 시계 방향으로 갈건지 시계 반대 방향으로 갈건지에 대한 것이다.
```

## 1.6 박스 그래프(Box Plot)

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

plt.style.use('default')

data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 2, 1500)

fig, ax = plt.subplots()

# 박스 그래프 : boxplot
ax.boxplot([data_a, data_b, data_c], notch = True, whis = 1.5, vert = True) # notch는 신뢰구간이 95%인 구간을 표시하는 것이다.
ax.set_ylim(-20.0, 20.0) # y의 최저값과 최고값을 표시하는 limit에 함축된 의미이다.
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

```python
# 가로 박스 그래프
fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c], notch = True, whis = 1.5, vert = False) # notch는 신뢰구간이 95%인 구간을 표시하는 것이다.
ax.set_xlim(-10.0, 10.0)
plt.show()
```

## 바이올린 플랫(Violin Plot)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')

# 데이터를 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 2, 1500)

# 바이올린 : plt.violinplot
violin = plt.violinplot([data_a, data_b, data_c], showmeans = True, 
                        showextrema = True, showmedians = True,) # showmeans는 중앙값을 표시할지이다.
plt.xticks(np.arange(1, 4), labels = ['A', 'B', 'C'])
plt.show()
```

* showmeans로 해서 평균값을 실행해준다.
* showmedians로 해서 중앙값을 실행해준다.
* showextrema로 해서 extrema를 실행해준다.

* 평균값과 중앙값이 조금 다르기 때문에 B에 있는게 가운데가 더 크다는 것을 볼 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')

# 데이터를 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 2, 1500)

# 바이올린 : plt.violinplot
violin = plt.violinplot([data_a, data_b, data_c], showmeans = True, 
                        showextrema = True, showmedians = True,) # showmeans는 중앙값을 표시할지이다.
plt.xticks(np.arange(1, 4), labels = ['A', 'B', 'C'])
violin['bodies'][0].set_facecolor('red')
violin['cbars'].set_edgecolor('red')
violin['cmaxes'].set_edgecolor('orange')
violin['cmins'].set_edgecolor('black')
violin['cmeans'].set_edgecolor('red')
violin['cmedians'].set_edgecolor('black')
plt.show()
```
