## 1. 확률

- 사건이 일어날 확률
- 고전적 확률 정의:
        - 실험결과가 N가지 일 때, 일어난 사건 A가 m번 일어났으면 P(A) = m/N
        - 각 실험 가능성은 모두 같을 때 
        - (예시 주사위 {1, 2, 3, 4, 5, 6} 1/6 1/6 1/6 1/6)
 
- 경험적 정의:
        - 반복 횟수를 N이라 했을 때 무한번 반복했을 때, 발생한 A 횟수 m
        - P(A) = lim(m/N)
        - (예시 주사위를 엄청많이 던져서 나오게 되는 확률)
 
- 공리적 정의:
        - 콜모코로프 공리적 정의 세가지 
        - 0 <= P(A) <= 1 이 사이에 있는 것이 만족한다.
        - P(S) = 1
        - 실험이 표본공간의 부분집합이면서 다른 사건과 교집합이 없을 때, 각 실험의 확률을 더한 값과 합집합의 확률은 같다.
        - 예를 들면, 동전 하나 던지기를 합니다.
        - 앞면: H, 뒷면: T
        - 표본공간 = S{HH, HT, TT, TH}
        - 앞면이 나오는 사건(A{TT})와 뒷면이 나올 사건(B{HH})면 A와 B는 교집합이 없다. 
        - P(A U B) = P({TT, HH}) = P({TT}) + P({HH})
        - P(A U B) = P(A) + P(B)

## 2. 확률의 성질
- 공집합의 확률 0
- 여집합의 확률 
        실험 A의 여집합은 1 - P(A)

- 포함-배제의 원리
        p(A U B) = P(A) + P(B) - P(A n B) 
        
- 각 사건이 배반사상(event)일 경우
        P(A1 U A2 U A3 ... U An) = P(A1) + P(A2) + ... + P(An)
        A가 B의 부분집합일 때, P(A) <= P(B)

## 3. 조건부 확률

- 어떤 사건 B가 일어날 때(P(B) > 0) 사건 A가 일어날 확률
$$P(A|B) = \frac{P(A\cap B)}{P(B)}$$


- 사건 A가 발생했을 때, 사건 B가 발생할 확률
$$P(B|A) = \frac{P(A\cap B)}{P(A)}$$


$$P(A\cap B) = P(A|B)P(B) = P(B|A)P(A)$$

- 독립사건
$$P(A|B) = P(A)$$
$$P(B|A) = P(B)$$

## 4. 베이즈 정리

- 분할 사건
        - 사건 K개가 일어났을 때 A1 ... AK 서로 다 교집합이 없는 경우 분할 사건
        - 사건 K개를 모두 합집합 했을 때, S가 되는 경우
        - 이 두가지 조건을 만족하는 경우 분할 사건이라고 한다.
$$ B = ({A_1}\cap {B}) U ... U ({A_K}\cap {B})$$

$$P(B) = P(B \cap {A_1}) + ⋅ ⋅ ⋅ + P(B ∩ {A_K})$$

$$P({A_i}|B) = \frac{P(B|{A_i})P({A_i})}{P(B)}$$

$$P({A_i}|B) = \frac{P(B|{A_i})P({A_i})}{P(B)} = \frac {P(B|{A_i})P({B_i})}{P(B|{A_1})P({A_1}) + P(B|{A_2})P({A_2})+ ⋅ ⋅ ⋅+ P(B|{A_k})P({A_k})}$$
         
         - 사전 확률을 사후확률로 추정하는 정리

## 5. 확률 변수와 확률 분포

- 확률 변수
        - 표본공간에서 정의된 실수값 함수
        - 확률 변수들의 통계량 찾고 
        - 확률 변수들을 통해서 모수를 추정
        - 대문자 X로 표현
- 확률 분포
        - 확률 변수 값과 확률을 대응시키는 표, 그래프, 함수
        
- 이산 확률 변수
        - countably infinite 변수(예 개수)
        - 확률 질량 함수를 통해 확률을 얻을 수 있다.
        
    - 확률 질량 함수
        $P(a <= X <= b) = \Sigma{f(x)}$

- 연속 확률 변수(예 키)
        - 하나로 특정할 수 없고 셀수 없는 변수
    - 확률 밀도 함수
        - 확률 밀도 함수 f(x)는 연속확률 변수 X에 대한 확률에 분포를 나타내는 함수
        - 모든 x에 대해서 f(x) >= 0
        - 확률 밀도 함수의 전체 면적은 1
        - 특정 구간에서의 확률은 확률밀도 함수를 적분한 값이다.

## 6. 정규분포

- 정의
       - 확률 변수 X가 평균이 m이고 분산이 r인 정규분포는 X ~ N(m, r)

- 특징
       - 모든 정규분포는 좌우 대칭
       - 정규곡선의 위치는 평균에 의해서 결정된다.
       - 표본공간에 따라 평균과 표준편차는 다르지만 평균으로부터 K배 표준편차 이내에 확률 변수 X가 존재할 확률은 동일

## 9. 분산과 표준편차

- 분산 확률분포함수의 모양을 설명할 수 있다.
- 분산은 확률분포함수에서 확률이 모여있는지 퍼져있는지를 알 수 있는 수치적 파라미터

$${\Sigma^2} = Var[X] = E[{(X - \mu)^2}]$$

- 성질:
$$Var[X] >= 0$$
$$Var[cX] = {c^2}Var[X]$$
$$Var[X + Y] = Var[X] + Var[Y] + 2E[(X - mx)(Y - my]$$

$$E[X] = E[\bar{X}]$$

* 표본평균의 분산과 표본의 분산의 관계
$$E[X] = E[\bar{X}]$$
$$Var[\bar{X}] = \frac{1}{N}Var[X]$$
