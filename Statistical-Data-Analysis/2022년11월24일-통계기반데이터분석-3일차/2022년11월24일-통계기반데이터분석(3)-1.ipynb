## 1. 다변수 확률변수

- 변수가 여러개일 때, 분포를 알아보고 연관된 분포 혹은 결과값들을 알아봅시다.
- 변수가 여러개가 되면, 확률분포가 결합되어 결합확률분포함수라고 합니다.
- 두개의 확률 변수 값의 결합확률분포를 알고 있을 때, 두 개 중에 하나의 확률 분포를 알고 있으면 나머지 하나의 확률분포도 알 수 있다.

## 1.1. 이산 데이터

- X, Y의 확률 질량 함수
       - X의 확률질량함수 = Px(x)
       - Y의 확률질량함수 = Py(y)
       - 결합 확률질량함수 = Pxy(x,y)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import scipy as sp
```

```python
# 학점에 대해 다루어 본다.
# 영어, 수학
grades = ['A', 'B', 'C', 'D', 'E', 'F']
scores = pd.DataFrame(
    [[1, 2, 1, 0, 0, 0],
     [0, 2, 3, 1, 0, 0],
     [0, 4, 7, 4, 1, 0],
     [0, 1, 4, 5, 4, 0],
     [0, 0, 1, 3, 2, 0],
     [0, 0, 0, 1, 2, 1]
    ],
    columns = grades, index = grades
)
scores.index.name = '수학'
scores.columns.name = '영어'
scores
```

```python
# 결합확률 질량함수로 나타내 본다.
scores.values.sum() 
```

* 총 학생 수 50명

```python
scores/scores.values.sum()
```

* 결합확률질량함수의 값

```python
import matplotlib.font_manager as fm # 폰트를 임포트 시켜준다.
for i in fm.findSystemFonts(fontpaths = None, fontext = 'ttf'):
    if 'nanum' in i.lower():
        print(i)
        
plt.rc('font', family = 'NanumGothic')
```

```python
# 결합확률질량함수 히트맵
pmf = scores/scores.values.sum()
sns.heatmap(pmf, annot=True, cmap = mpl.cm.bone_r, xticklabels = grades, yticklabels = grades)
plt.title("결합확률질량함수 pxy(x,y)")
plt.tight_layout()
plt.show()
```

## 1.2. 주변확률질량함수

- 결합확률 질량함수에서 단변수 확률질량함수를 구하는 것.

$${px}(x) = \Sigma{Pxy}(x,{y_i})$$

$${py}(y) = \Sigma{Pxy}({x_i}, y)$$

```python
# 결합확률질량함수에서 주변확률질량함수의 영어에 대한 확률질량함수
x_english = pmf.sum()
```

```python
# 결합확률질량함수에서 주변확률질량함수의 수학에 대한 확률질량함수
y_math = pmf.sum(axis = 1)[:,np.newaxis]
```

## 1.3. 조건부확률질량함수

X,Y 변수 중 하나의 변수가 특정 값으로 고정되는 경우 나머지 하나의 변수의 확률질량함수를 구하는것

$$P_{X|Y}(x|y)=\frac{P_{XY}(x,y)}{P_Y(y)}$$

$$P_{Y|X}(y|x)=\frac{P_{XY}(x,y)}{P_X(x)}$$

# 2. 다변수 연속확률 변수

## 2.1. 결합누적확률분포함수
이산확률분포처럼 단순사건을 이용할 수 없음(A,B,C,D)   
키, 몸무게, 속도, 시간, 온도   
CDF -> 미분 -> PDF   
결합누적확률분포함수 -> 미분 -> 결합확률밀도함수

$${F_{XY}} = P({X<x, Y<y})$$

- 특성
$${F_{XY}}(-∞, ∞) = 1$$
$${F_{XY}}(-∞, y) = {F_{XY}}(x, ∞) = 0$$

## 2.2. 결합확률 밀도 함수
결합누적확률분포함수를 미분해서 정의

$${P_{XY}}(x,y)$$

- 특성
      적분하면 1
$$ ∫∫{P_{XY}}(x,y)dxdy = 1 $$

```python
mu = [70, 170] # 몸무게, 키
cov = [
    [150, 140],
    [140, 300]
]
'''
공분산 행렬 : [
  [분산, 공분산],
  [공분산, 분산]]
]
'''
rv = sp.stats.multivariate_normal(mu, cov) # 몸두게와 키에대한 다변수 정규분포이다.
x, y = np.linspace(20, 120, 100), np.linspace(100, 250, 100)
```

```python
xm, ym = np.meshgrid(x, y)
z = rv.pdf(np.dstack([xm, ym]))
```

* dstack은 두개의 어레이를 쌓는 것이다.

```python
z.sum()
```

```python
plt.contour(x, y, z) 
plt.xlabel('x')
plt.ylabel('y')
plt.title('결합확률밀도함수 P(x, y)')
plt.show()
```

* 색상이 연한 초록색으로 될 수록 밀도가 높다.

## 2.3. 주변확률밀도함수

- 주변확률밀도 함수는 하나의 변수를 가중평균을 하고 나머지 하나의 변수에 대해서 확률밀도함수를 구하는 것.
- 주변확률밀도함수는 결합확률 밀도 함수를 통해 단일변수의 확률밀도함수를 구하는 것.

$${P_X}(x) = ∫{P_{XY}}(x,y)dy$$

$${P_Y}(y) = ∫{P_{XY}}(x,y)dx$$

```python
from matplotlib.ticker import NullFormatter
from matplotlib import transforms
from scipy.integrate import simps # 적분을 하기위해 사용하는 라이브러리이다.

mu = [70, 170] # 몸무게, 키
cov = [
    [150, 140],
    [140, 300]
]

x = np.linspace(20, 120, 100)
y = np.linspace(100, 250, 100)

xm, ym = np.meshgrid(x, y)
rv = sp.stats.multivariate_normal(mu, cov)
z = rv.pdf(np.dstack([xm, ym]))

fx = [simps(z1, y) for z1 in z.T] # 여기서 T는 전치
fy = [simps(z1, x) for z1 in z]

left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.05

rect1 = [left, bottom, width, height] # 사각형의 정보를 넣어주기 위해서 뭉치는 것이다.
rect2 = [left, bottom_h, width, 0.2]
rect3 = [left_h, bottom, 0.2, height]

ax1 = plt.axes(rect1)
ax2 = plt.axes(rect2)
ax3 = plt.axes(rect3)

ax2.xaxis.set_major_formatter(NullFormatter())
ax3.xaxis.set_major_formatter(NullFormatter())

ax1.contour(xm, ym, z)
ax1.set_title('결합확률분포 함수')
ax1.set_xlabel('x')
ax1.set_xlabel('y')

ax2.plot(x, fx)
ax2.set_title('주변확률밀도함수 x(몸무게)')

base = ax3.transData
rot = transforms.Affine2D().rotate_deg(-90) # 아핀을 통해서 90도 회전을 시켜주었다.
plt.plot(-y, fy, transform = rot + base)
plt.title('주변확률밀도함수 y(키)')

ax1.set_xlim(38, 102)
ax2.set_xlim(38, 102)
ax3.set_xlim(0, 0.025)

ax1.set_ylim(120, 220)
ax3.set_ylim(120, 220)
```

## 2.4 조건부 확률 밀도함수

- 다 변수 확률 변수 중에 하나가 특정한 값일 경우(연속형은 특정한 값을 지정할 수 없지만 가정을 한다.) 나머지 확률밀도함수를 조건부 확률 밀도함수

$$P_{X|Y}(x|y)=\frac{P_{XY}(x,y)}{P_Y(y)}$$

$$P_{Y|X}(y|x)=\frac{P_{XY}(x,y)}{P_X(x)}$$

```python
from matplotlib.collections import PolyCollection
from matplotlib import colors as mcolors

mu = [70, 170] # 몸무게, 키
cov = [
    [150, 140],
    [140, 300]
]

x = np.linspace(20, 120, 100)
y = np.linspace(100, 250, 16)  # 컴퓨터상에서는 연속해서 구할수 없기 때문에 그 중에서 16개만 뽑아서 구해보았다.

xm, ym = np.meshgrid(x, y)
rv = sp.stats.multivariate_normal(mu, cov)
z = rv.pdf(np.dstack([xm, ym]))

fig = plt.figure(dpi = 150) # 1인치 안에 들어가는 점의 개수를 dpi로 설정해준다.
ax = fig.gca(projection = '3d')

xs = np.hstack([0, x, 0])
zs = np.zeros_like(xs)
verts = []

for i, y1 in enumerate(y):
    zs[1:-1] = z[i]
    verts.append(list(zip(x, zs)))
    
poly = PolyCollection(verts)
poly.set_alpha(0.5)
ax.add_collection3d(poly, zs = y, zdir = 'y')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(20, 120)
ax.set_ylim(100, 250)
ax.set_zlim3d(0, 0.0007)
ax.view_init(50, -50)
plt.title('결합확률밀도함수의 단면')
plt.show()
```

```python
mu = [70, 170] # 몸무게, 키
cov = [
    [150, 140],
    [140, 300]
]

n = 10
x = np.linspace(20, 120, 100)
y = np.linspace(100, 250, 16)

xm, ym = np.meshgrid(x, y)
rv = sp.stats.multivariate_normal(mu, cov)
z = rv.pdf(np.dstack([xm, ym]))

plt.figure(figsize=(8,6))

for i, j in enumerate(range(5, 10)):
  ax = plt.subplot(5, 1, i+1)  # 5개 중 하나인데 지정을 해준다.
  ax.tick_params(labelleft=False)
  plt.plot(x, z[j,:] * 10, label='결합확률밀도함수의 단면')
  density = simps(z[j,:], x)
  plt.plot(x, z[j,:]/density, label='조건부확률밀도함수') # y를 나눠주는 것을 density라고 표현하였다.
  ax.xaxis.set_ticklabels([])
  plt.ylabel(f'p(x, y={y[j]}', rotation=0, labelpad=30)
  if i == 0:
    plt.legend(loc=2)
plt.xlabel('x')
plt.tight_layout
```

## 독립

- 표본 값이 달라질 때 다른 확률 변수의 조건부 분포가 달라질 때 상관관계가 있다.
- 두 확률변수가 상관관계가 아니면 서로 독립이다.

$${P_{XY}}(x,y) = {P_X}(x){P_Y}(Y)$$

## 조건부 확률분포

- 독립인 X, Y의 조건부 확률밀도함수는 주변확률밀도함수와 같다.

$${P_{X|Y}}(x|y) = \frac{{P_X}(x){P_Y}(y)}{P_Y(y)} = {P_X(x)}$$
$${P_{Y|X}}(y|x) = \frac{{P_X}(x){P_Y}(y)}{P_X(x)} = {P_Y(y)}$$

```python
# 이산 변수에 대한 예제
pmf1 = np.array(
    [
    [1, 2, 4, 2, 1],
    [2, 4, 8, 4, 2],
    [4, 8, 16, 8, 4],
    [2, 4, 8, 4, 2],
    [1, 2, 4, 2, 1]
    ]
)
pmf1 = pmf1/pmf1.sum()
```

```python
pmf1
```

```python
pmf_x = pmf1.sum(axis = 0)
pmf_y = pmf1.sum(axis = 1)
```

```python
pmf_x
```

```python
pmf_y
```

```python
pmf1x = pmf_x * pmf_y[:, np.newaxis] # 서로 곱하기 위해서 새로운 축을 가지고 온다.
```

```python
plt.subplot(121)
sns.heatmap(pmf1, annot = True, square = True, cbar = False, xticklabels = pmf_x, yticklabels = pmf_y)
plt.title('독립인 두 확률변수의 결합확률질량 함수')

plt.subplot(122)
sns.heatmap(pmf1x, annot = True, square = True, cbar = False, xticklabels = pmf_x, yticklabels = pmf_y)
plt.title('독립인 두 확률변수의 결합확률질량 함수의 곱')
plt.tight_layout()
plt.show()
```

* x와 y는 서로 독립이라는 것을 알 수 있다.

## 상관관계가 있는 두 확률변수

```python
pmf2 = np.array([
    [0, 0, 0, 5, 5],
    [0, 5, 5, 5, 5],
    [0, 5, 30, 5, 0],
    [5, 5, 5, 5, 0],
    [5, 5, 0, 0, 0],
])
pmf2 = pmf2/pmf2.sum()

pmf2_x = pmf2.sum(axis = 0)
pmf2_y = pmf2.sum(axis = 1)
pmf2x = pmf2_x * pmf2_y[:,np.newaxis]

plt.subplot(121)
sns.heatmap(pmf2, annot = True, square = True, cbar = False, xticklabels = pmf_x, yticklabels = pmf_y)
plt.title('상관관계가 있는 두 확률변수의 결합확률질량 함수')

plt.subplot(122)
sns.heatmap(pmf2x, annot = True, square = True, cbar = False, xticklabels = pmf_x, yticklabels = pmf_y)
plt.title('두 변수의 주변확률질량함수의 곱')
plt.tight_layout()
plt.show()
```

* x와 y가 서로 상관관계가 있다는 것을 히트맵으로 알 수 있다.
