## 1. 선형회귀분석 기초

$$\hat{y} = f(x)$$
$$\hat{y} = w_0 + w_1x_1 + w_2x_2 + \cdots + w_Dx_D$$
$$= w_0 + w^Tx$$

### 1.1 최소자승법(OLS)

y와 $\hat{y}$이 가까워 지려면 OLS(Oridnary Least Squares)라는 방법을 통해 RSS(Residual Sum of Squares)를 최소화시키는 것과 같다.

$$\hat{y} = Xw$$
$$ϵ=y-\hat{y} =y - Xw$$
$$RSS = ϵ^Tϵ=(y - Xw)^T(y - Xw)$$
$$ = y^T-2y^TXw + w^TX^TXw$$ 
* RSS(잔차제곱합)을 w로 미분하여 잔차제곱합의 그레디언트 벡터를 구합니다. 잔차가 최소가 되기 위해서는 그레디언트 벡터가 0이어야 합니다.
* 그래디언트가 0이 되도록 하는게 제곱법이 최소가 된다.

$$\frac{dRSS} {dw} = 0 $$
$$\frac{dRss} {dw} = -2X^Ty + 2X^TXw$$
$$X^Ty = X^TXw$$
$$w = (X^TX)^{-1}X^{T}y$$

* 이제 w(가중치)를 찾기 위한 방법을 취한다.
* $(X^TX)^{-1}$이 있을 때 역행렬을 구할 수 있다.

### 1.2 직교 방정식
그레디언트 벡터가 0벡터가 되는 관계를 직교방정식

$$X^Ty - X^TXw = 0$$
$$X^T(y - Xw) = 0$$
$$X^Te = 0$$

- 직교 방정식의 성질
        (1). 잔차 벡터의 원소의 합은 0
$$\Sigma_{i = 0}^Ne_i = 0$$

        (2). x  데이터의 평균값 xbar에 대한 예측값은 y 데이터의 평균값이다.
        

```python
from sklearn.datasets import make_regression
import statsmodels.api as sm

bias = 100
X0, y, w = make_regression(
    n_samples = 200, n_features = 1, bias = bias, noise = 10, coef = True, random_state = 1
)
X = sm.add_constant(X0)
y = y.reshape(len(y), 1)
```

```python
w
```

* y = 100 _86.447 x + 에타 로 되어 있는 것을 알 수 있다.

```python
# OLS 해 구하기
w = np.linalg.inv(X.T @ X) @ X.T@y # @는 행렬곱이다.
w
```

* OLS로 해를 구하니깐, $\hat{y}$ = 99.79 + 86.96 $x$로 되어 있는 것을 알 수 있다.

```python
# 예측을 생성하기 위한 데이터를 생성하는 것이다.
x_new = np.linspace(np.min(X0), np.max(X0), 10)
X_new = sm.add_constant(x_new)
y_new = np.dot(X_new, w)

plt.scatter(X0, y, label = '기존데이터')
plt.plot(x_new, y_new, 'rs-', label = '예측값')
plt.title('선형 회귀분석')
plt.legend()
plt.show()
```

* OLS로 RSS가 0이 되는 값을 찾았을 때, w를 찾고 x와 w와 잔차를 더하고 y를 예측하였다.

## 범주형 독립변수

 - D개의 독립변수를 가지는 선형회귀모델
$$\hat{Y} = w_0 + w_1x_1 + ⋯ + w_Dx_D$$

$$\hat{Y} = w_{A, 0} + w_{A, 1}x_2 + ⋯ + w_{A, D} x_{D}$$
* A인 것만 가지고 와서 계산을 해주면 A에 대한것만 가지고 계산을 해준다.

$$\hat{Y} = w_{B, 0} + w_{B, 1}x_2 + ⋯ + w_{B, D} x_{D}$$
* B인 것만 가지고 와서 계산을 해주면 B에 대한것만 가지고 계산을 해준다.

* 가중치들이 다르기 때문에 변수를 바꾸어 준다. 그래서 더미변수를 사용해준다.


- 더미 변수 방법
    - 풀랭크 방식
$$w_1x_1 = w_{1A}d_{1A} + w_{1B}d_{1B}$$
$$\hat{Y} = w_0 + w_{1A}d_{1A}+w_{1B}d_{1B} + ⋯ + w_Dx_D$$
* 하나를 없애줌으로써 하나의 수식을 만들 수 있다.

$$x1 = A -> d_{1A} = 1, d_{1B} = 0$$
$$x1 = B -> d_{1A} = 1, d_{1B} = 1$$

$$x_1 = A : \hat{Y} = w_0 + w_{1A}+ w_2x_2 + ⋯ + w_Dx_D$$
$$x_1 = B : \hat{Y} = w_0 + w_{1B}+ w_2x_2 + ⋯ + w_Dx_D$$

```python
import datetime
from calendar import isleap

def convert_partial_year(number): # <- 전처리를 해준다.
    # 날짜 계산
    year = int(number)
    d = datetime.timedelta(days = (number - year) * (365 + isleap(year)))
    day_one = datetime.datetime(year, 1, 1)
    date = d + day_one
    return date

df_nottem = sm.datasets.get_rdataset('nottem').data
```

```python
df_nottem.head()
```

```python
df_nottem['date0'] = df_nottem[['time']].applymap(convert_partial_year) # time이 float로 되어 있는 것을 월별로 바꾸어준다.
df_nottem['date'] = pd.DatetimeIndex(df_nottem['date0'].round('60min') + datetime.timedelta(seconds = 3600 * 24))
df_nottem['month'] = df_nottem['date'].dt.strftime('%m').astype('category')

del df_nottem['date0'], df_nottem['date']
df_nottem.head()
```

```python
df_nottem.info() # 카테고리 값으로 변경이 되어 있는 것을 알 수 있다.
```

```python
# 각각의 월별로 온도에 대해서 박스플롯을 그려본다.
df_nottem.boxplot('value', 'month')
plt.show()
```

```python
# x : 월 (12개) <- 독립변수
# 종속변수 (y) : 월 평균기온
model = sm.OLS.from_formula('value ~ C(month)', data = df_nottem) # C는 해당값을 카테고리라고 명시시켜주는 얘이다.
result = model.fit()
print(result.summary())
```
