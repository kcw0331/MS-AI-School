## 베르누이 분포
        bernoulli
        베르누이 실험 : 결과가 두개 밖에 없는 실험 (예. 동전)
        베르누이의 확률변수 : 이산확률변수 {0, 1}, {-1, 1} 0이나 1로 표현하거나 -1이나 1로 표현하는 경우도 있다.
$$Bern(x;\mu)={\mu^x}{(1-\mu)^{(1-x)}}$$
        * 여기에서 뮤는 1이거나 1이아니거나 하는 확률이다.

```python
mu = 0.6
rv = scipy.stats.bernoulli(mu)
rv
```

```python
x = [0, 1]
plt.bar(x, rv.pmf(x))
plt.xticks([0, 1], ['x=0', 'x=1'])
plt.grid()
plt.xlabel('표본값')
plt.title('베르누이 확률질량 함수')
plt.show()
```

* 동전이 앞면이 나올 확률이 0.6이니깐, 동전이 뒷면이 나올 확률은 0.4라는 것을 알 수 있다.

```python
x = rv.rvs(100, random_state = 0)
x
```

* 앞면이 나오거나 뒷면이 나올 경우 밖에 없으니깐, 0과 1로만 표시되는 것을 볼 수 있다.

```python
sns.countplot(x) # 횟수를 그래프로 그려주는 것이다.
plt.title('베르누이의 확률변수 시뮬레이션')
plt.grid()
plt.show()
```

* 표본의 수에 따라 이론 값과 시뮬레이션의 차이에 대해 알아본다.

```python
y = np.bincount(x, minlength = 2)/float(len(x))

df = pd.DataFrame({'이론' : rv.pmf([0, 1]), '시뮬레이션' : y})
df.index = [0, 1]
df
```

* 실제로는 0이 나올 확률은 0.38이고 1이 나올 확률은 0.62라는 것을 시뮬레이션을 통해서 볼 수 있다.

```python
for i in range(200):
    x = rv.rvs(i, random_state = 0)
    y = np.bincount(x, minlength = 2)/float(len(x))
    print(f'표본수가 {i}개일 때 이론값 : {rv.pmf([0, 1])} 표본값 : {y}')
```

## 이항분포

- 베르누이 분포에서 실험을 여러번 한 것이 이항분포

$$X\sim Bin(x;N,\mu)$$

표본 중 성공일 경우 1 아닐 경우 0

N개의 표본 중 성공 횟수는 N개의 표본을 모두 더한 것과 같다.
$$x=\Sigma{y_i}$$

확률 질량 함수
$$Bin(x;N,μ) = \binom{N}{x}{μ^x}{(1-μ)^{(N-x)}}$$

```python
N = 10
mu = 0.6
rv = scipy.stats.binom(N, mu) # 이항 분포에서는 N을 넣어준다.
```

```python
x_ = np.arange(N + 1)
plt.bar(x_, rv.pmf(x_), align = 'center')
plt.title('이항분포의 확률질량함수')
plt.xlabel('표본값')
plt.ylabel('P(X_)')
plt.show()
```

* 실험 횟수에 따라 성공 확률을 나타내고 있다. 그래서 이걸 다 더하게 되면 1이 된다.
* 성공율이 0.6이기 때문에 6일때가 가장 높다는 것을 알 수 있다.

- 보통 자연현상에서 나타나는 수치 정규분포 가정을 많이 함

$$N(x;\mu, {q^2}) = \frac{1}{(2\pi{\sigma^2})^\frac{1}{2}}exp(-\frac{({x-\mu})^2}{2{\sigma^2}})$$

## Q-Q 플롯

- 어떤 확률변수의 분포가 정규분포인지 검사하는 그래프

- 정규분포를 따를 때 Q-Q 플롯은 직선의 형태를 가집니다.

- 정규분포를 따르지 않을 때 Q-Q 플롯은 곡선이 발생

```python
mu = 0
std = 1
rv = scipy.stats.norm(mu, std)
```

```python
x = rv.rvs(100)
x_sorted = np.sort(x)
x_sorted
```

```python
from scipy.stats.morestats import _calc_uniform_order_statistic_medians
```

```python
position = _calc_uniform_order_statistic_medians(len(x))
```

```python
position
```

```python
ff = rv.ppf(position)
ff
```

* x를 넣어서 분위수를 찾았다.

```python
plt.scatter(ff, x_sorted)
plt.title('QQ')
plt.axis('equal')
plt.show()
```

* 현재 보이는 이것이 정규분포를 따른다는 것을 알 수 있다.

```python
# 이방법을 사용하면 쉽게 구할 수 있다.

np.random.seed(0)
plt.figure(figsize=(7,7))
sp.stats.probplot(x, plot=plt)
plt.axis('equal')
plt.show()
```

```python
# 정규분포를 따르지 않는 분포는 어떻게 되는지 QQ플롯을 그려와 준다.

np.random.seed(0)
x = np.random.rand(100)
plt.figure(figsize = (7, 7))
scipy.stats.probplot(x, plot=plt) # plt객체를 출력해주어야 한다.
plt.show()
```

* 곡선이 그려지게 되면 정규분포를 따르지 않는다는 것을 알 수 있다.

```python
petal_width = load_iris().data[:, 3]
```

```python
petal_width
```

## 중심 극한 정리

- 여러개의 확률변수의 합이 정규분포와 비슷한 분포를 이루는 현상
- 확률 변수의 평균들의 평균은 확률변수의 수가 무한히 많아지게 되면, 모집단의 평균과 분산이 된다.

```python
np.random.seed(0)
xx = np.linspace(-2, 2, 100)

plt.figure(figsize = (6, 9))

for i, N in enumerate([1, 2, 10]):
    X = np.random.rand(5000, N) # 표본하나당 5000개를 가지고 온다. 5000개의 표본 데이터를 가지고 있는 하나를 가지고 온다.
    Xbar = (X.mean(axis = 1) - 0.5) * np.sqrt(12 * N) # 평균을 가지고 온다.
    ax = plt.subplot(3,2, 2 * i + 1)
    sns.distplot(Xbar, bins = 10, kde = False, norm_hist = True) # 10개의 구간을 나누어서 본다.
    ax.set_title(f'N = {N}') # N이 표본의 수이다.
    plt.subplot(3, 2, 2 * i + 2)
    sp.stats.probplot(Xbar, plot = plt)
    
plt.tight_layout()
plt.show()
```

* 표본이 하나일 때, 랜덤하게 데이터를 가지고 왔을 때 균일 분포를 이룬다.
* 표본이 많아지게 되면, 정규분포를 따르게 된다. -> 중심극한 정리

## 스튜던트 t 분포

- 정규분포에서 양끝단의 분포가 많아지는 분포를 t분포(fat tail 현상)

- t분포의 기댓값
        정규분포와 동일 : 평균
        
- t분포의 분산
$$Var[X]=\frac{v}{λ(v-2)}$$

```python
import pandas_datareader.data as web

symbols = ['SP500','NASDAQCOM','DJCA', 'NIKKEI225']
data = pd.DataFrame()
for sym in symbols:
  data[sym] = web.DataReader(sym, data_source='fred')[sym]
data = data.dropna()
```

```python
(data/data.iloc[0]*100).plot() # 증감 추이를 보이기 위해서 나누어 주었다.
plt.xlabel('날짜')
plt.ylabel('추가 수익율')
plt.grid()
plt.show()
```

```python
log_returns = np.log(data/data.shift(1))
log_returns.hist(bins = 50)
plt.tight_layout()
plt.show()
```

```python
for i, sym in enumerate(symbols):
    ax = plt.subplot(2, 2, i + 1)
    sp.stats.probplot(log_returns[sym].dropna(), plot = ax)
plt.tight_layout()
plt.show()
```

* t 분포에서는 2이상의 자유도를 가진다.

```python
xx = np.linspace(-4, 4, 100)
for df in [1, 2, 5, 10, 20]:
    rv = sp.stats.t(df = df)
    plt.plot(xx, rv.pdf(xx), label = f'스튜던트 t 분포(df = {df})')
plt.plot(xx, sp.stats.norm().pdf(xx), label = '정규분포', lw = 5, alpha = 0.5)
plt.title('자유도에 따른 스튜던트 t분포의 변화')
plt.ylabel('P(x)')
plt.legend()
plt.show()
```

* t 분포와 같은 경우 정규분포와 닮았는데, 자유도가 낮을 수록 이러한 첨도를 가지는 것을 알 수 있다.
* t 분포의 기대값은 뮤이고 분산은 자유도에 따라서 비례하다.

## 카이제곱 분포

- t분포와는 다르게 제곱을 해서 양수값만 가지도록 만든 것이 카이제곱 분포이다.

- 얼마나 특별한 사건이 일어나는지를 확인하는 분포

- t분포와는 다르게 제곱을해서 양수값만 가지도록 만든것이 카이제곱 분포

- 표본 분포가 모분포를 잘 반영하는지 알아낼 때 사용하는 분ㄴ포

```python
xx = np.linspace(0.01, 10, 100)
dfs = np.arange(1, 5)
lss = ['-', '--', '-.', ':']

for df, ls in zip(dfs, lss):
    rv = scipy.stats.chi2(df = df)
    plt.plot(xx, rv.pdf(xx), ls = ls, label = f"자유도 : {df}")
plt.title('자유도에 따른 카이제곱 분포')
plt.ylim(0, 0.6)
plt.xlabel('표본값')
plt.ylabel('P(X)')
plt.legend()
plt.show()
```

* 자유도에 따라서 분포가 이루어진다는 것을 알 수 있다. 
* 자유도가 1, 2, 3, 4에 따라 다른 양상을 보이는 것을 알 수 있다. 

### F-분포

- 여러 집단에 대해서 각 집단의 평균이 의미있게 차이가 나는지 측정 할 때 사용됩니다.   
- 선형 회귀 분석에 회귀분석모델이 전체 데이터 변동을 잘 설명하고 있는지 비교하기 위해서 사용됩니다.

```python
xx = np.linspace(0.03, 3, 1000)
plt.plot(xx, scipy.stats.f(1,2).pdf(xx), ls = '-', label='F(1,2)')
plt.plot(xx, scipy.stats.f(3,5).pdf(xx), ls = '-', label='F(1,2)')
plt.plot(xx, scipy.stats.f(10,10).pdf(xx), ls = '-', label='F(1,2)')
plt.title('자유도에 따른 F분포의 모양')
plt.legend()
plt.show()
```

## 확률분포의 결정

확률변수가 베르누이, 이항, 정규분포 중에 어떤 확률 분포를 따르는지 알아내고 표본집단의 확률분포의 모수값을 구하는 작업을 합니다.

- 데이터가 0이나 1뿐이면 -> 베르누이 분포를 따른다고 할 수 있다.
- 데이터가 카테고리 값을 경우 -> 카테고리 분포
- 데이터가 0과 1사이에 실수 값일 경우 -> 베타 분포
- 데이터가 항상 0이상일 경우 -> 로그정규분포, 감마분포, f분포, 카이제곱분포, 지수 등등 
- 데이터가 제한이없고 연속된 실수일 경우 -> 정규분포, 스튜던트 t-분포, 라플라스 분포 등등 

데이터의 특성을 알면 어떤 분포를 따를지 추측이 가능하다.

하지만 꼭 이렇다고 이러한 분포를 가진다는 것은 아니라는 것을 명심하자.!
