```python
from azure.eventhub.aio import EventHubProducerClient
from azure.eventhub import EventData  
```

* 두개의 패키지를 불러온다.

```python
eventhub = EventHubProducerClient.from_connection_string(conn_str = '') # 연결 문자열을 준다. 
```

* Azure 포털에서 Event Hubs를 생성해주고 그 안에서 Hubs안에 Hub를 만들어 준다음에 Shared access policies에서 Add해서 me를 생성해주고 Connection-string-primary key를 복사해서 conn_str에 붙여주었다.

```python
import time # 일정 주기로 시간을 보낸다.
from datetime import datetime # 시간을 보내준다.

i = 0 # 반복하는 횟수를 보기위해서 
while(1):  # 무한 루프를 만들어준다.
  message = 'time: ' + str(datetime.now()) # 현재 시간 뒤에 now를 붙여준다.
  eventhub.send_event(EventData(message)) # eventhub로 전송을 하게 된다. , send 배치(batch)는 한꺼번에 무언가를 보낼 때 사용한다. EventHub에 적합한 형태로 이벤트를 보낸다.
  
  i += 1
  if i % 10 == 0: # 20번에 한번씩 메세지를 뿌려준다.
    print('message {} sending...'.format(i)) # i를 출력하게 끔 해준다.
  time.sleep(2) # 2초에 한번씩 쉬어가면서 보낸다.
```

* 10초에 한번씩 데이터를 쏜다고 볼 수 있다.
