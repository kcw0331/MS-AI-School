```python
# File Share 연결하는 코드

from azure.storage.file import FileService
file_service = FileService(account_name = '', # account 이름과 키가 필요하다.
                           account_key = '')
# account 키는 제거해준다.
```

```python
# myshare 공유 폴더를 생성
file_service.create_share('myshare') # myshare로 폴더를 만들어 준다.
```

* True라고 나오면 잘 실행 된 것이다.

```python
# myshare 안에 샘플 디렉토리인 sampledir를 생성한다.
file_service.create_directory('myshare', 'sampledir') # 이렇게 하면 샘플 디렉토리를 만들 수 있다.
```

* 그러면 디렉토리가 생긴 것을 볼 수 있다.

```python
# dog.jpg를 myfile로 해당 공유 폴더에 복사한다.

from azure.storage.file import ContentSettings # 이 패키지를 생성한다. 파일의 종류에 따라서 다루는 방법이 조금씩 다르다.

file_service.create_file_from_path('myshare', # 이 경로에 있는 파일을 사용하겠다는 것이다.
                                   None, 
                                   'myfile', # myfile로 저장이 된다.
                                   'dog.jpg', # dog이미지를 myfile로 저장을 시켜준다.
                                   content_settings = ContentSettings(content_type='image/jpg')) # 이미지 타입을 알려준다.  
```

* dog이미지를 업로드 해준다.
* 그럼 애저의 스토리지 계정의 파일공유에서 dog의 이미지가 myfile로 저장된 것을 볼 수 있다.

```python
# 공유 폴더안에 파일 목록을 가져와서 보여준다.

generator = file_service.list_directories_and_files('myshare')  # 여기에서 우리가 myshare라고 하면 파일의 경로를 가지고 올 수 있다.
for file_or_dir in generator:
  print(file_or_dir) # 그럼 파일의 목록이 나오는 것을 확인할 수 있다.
  print(file_or_dir.name)
```

* 우리가 파일을 올리는 것 까지 해 보았다.

```python
# 다시 myfile로 복사된 dog.jpg를 out-dog.jpg로 가져온다.
file_service.get_file_to_path('myshare', None, 'myfile', 'out-dog.jpg') # 특정 경로에서 파일을 가지고 와라 하는 거다.
```

* 파일을 올렸던 것을 가지고 오는 것을 해보았다.
