# 텐서

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

```python
x = np.array(3)
print(x)
print(x.shape)  # 0차원은 모양이 없다.
print(np.ndim(x)) # 차원이 없기 때문에 0이 되서 나온다.
```

* 스칼라의 값은 방향성이 없다.

```python
# 벡터 (1차원 텐서)
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
c = a + b

print(c)  # 같은 차원끼리 덧셈이 된다.
print(c.shape)
print(np.ndim(c))
```

```python
# 벡터의 곱

c = a * b
print(c)
print(c.shape)
print(np.ndim(c))
```

* 1차원으로 나온 것을 알 수 있다.

```python
# 스칼라와 벡터의 곱

a = np.array(10) # 스칼라
b = np.array([1, 2, 3]) # 1차원 텐서
c = a * b

print(c)
print(c.shape)
print(np.ndim(c))
```

```python
# 전치행렬

A = np.array([[1, 2, 3], [4, 5, 6]])
print('A\n', A)
print('A.shape\n', A.shape)
print('------------------')
```

* 전치행렬 전 

```python
# 전치행렬

A_ = A.T  # T가 속성으로 들어와 있다.
print('A_\n', A_)
print('A_.shape\n', A_.shape)
print('------------------')
```

* 속성 T를 사용하고 전치행렬 후

```python
# 번외편 PKL 파일 저장 및 로딩

from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import joblib

dataset = datasets.load_iris()

X, y = dataset['data'], dataset['target']

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

joblib.dump(model, './knn_model.pkl')
```

```python
loaded_model = joblib.load('./knn_model.pkl')
loaded_model.predict
score = loaded_model.score(X, y)
print('정확도: {score:.3f}'.format(score = score))
```
