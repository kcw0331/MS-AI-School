# Supervised Learning - Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
data = iris.data
label = iris.target
columns = iris.feature_names
```

```python
data = pd.DataFrame(data, columns = columns)
data.head()
```

```python
data.shape
```

## 1. Logistic Regression

```python
# 데이터의 준비
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, label, test_size = 0.2, random_state = 2022)

# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score   # classification에서 가장 중요한 것은 정확도!!!
```

```python
print('로지스틱 회귀, 정확도: {:.2f}'.format(accuracy_score(y_test, y_pred)))
```

## 2. Support Vector Machine

```python
# 서포트 벡터
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

print('서포트 벡터 머신, 정확도: {:.2f}'.format(accuracy_score(y_test, y_pred)))
```

* SVC의 C값을 크게 해주면 좋을 것 같지만 분류는 데이터에 따라 다르다.

## 3. Decision Tree

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth = 5)

dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)

print('결정 트리, 정확도: {:.2f}'.format(accuracy_score(y_test, y_pred)))
```

## 4. Random Forest

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth = 5)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print('랜덤 포레스트, 정확도: {:.2f}'.format(accuracy_score(y_test, y_pred)))
```
