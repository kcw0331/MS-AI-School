# Conditional

```python
import requests  # requests는 웹에다가 요청을 해서 데이터를 받아온다.
```

```python
response = requests.get("http://www.naver.com") # requests를 사용해서 네이버 웹사이트의 응답에 대해서 받을 수 있다.
```

* 통신이 잘 되면 200이 나오고
* 통신이 오류가 나면 400이 나오고
* 통신이 에러가 나면 500이 나오게 된다.

* requests에 대한 이야기를 해주셨다.

```python
a = 1
b = 1

if a == b:
    print('a와 b의 값이 같습니다.')
if a != b:
    print('a와 b의 값이 서로 다릅니다.')
```

```python
a = 1
b = 1

if a == b:
    print('a와 b의 값이 같습니다.')
else:
    print('a와 b의 값이 서로 다릅니다.')
```

```python
a = 1
b = 2

if a == b:
    print('a와 b의 값이 같습니다.')
elif a < b:
    print('a는 b 보다 작습니다.')
else:
    print('a는 b보다 큽니다.')
```
