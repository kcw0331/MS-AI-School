# Unsupervised Learning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
print(wine.DESCR)
```

```python
data = wine.data
label = wine.target
columns = wine.feature_names
```

```python
data = pd.DataFrame(data, columns = columns)
data.head()
```

## 1. K-Means
 - 점의 개수가 K이고, 평균값을 찾는다는 것이다.

```python
# K-means

# 데이터 전처리
# MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = scaler.fit_transform(data) 
```

```python
data.shape
```

* 13차원의 데이터다 라고 할 수 있다.

```python
# 13차원이기 때문에 패턴을 찾아내기 위해서 차원을 축소해준다.
# PCA (차원의 축소)
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
data = pca.fit_transform(data)
```

```python
data  # 2개의 컬럼이 13개의 특징을 나타내준다. (2차원)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)  # 와인은 3종류이기 때문에 3종류라고 지정해준다.
```

* KMeans 패키지를 사용한다.

```python
kmeans.fit(data)  # 데이터를 학습시켜 준다.
```

```python
cluster = kmeans.predict(data)  # 데이터를 유추해본다.
```

```python
type(cluster)  # 유추한 결과는 넘파일 배열로 나오게 된다.
```

```python
cluster
```

```python
# 데이터 시각화 
plt.scatter(data[:, 0], data[:, 1], c = cluster
            , edgecolor = 'black', linewidth = 1)  # x좌표는 data[:, 0], y좌표는 data[:, 1]
```

## 2. Hierarchical Clustering


```python
# Hierachical Clustering
from sklearn.cluster import AgglomerativeClustering
single_clustering = AgglomerativeClustering(n_clusters = 3, linkage = 'single') # linkage 종류를 3가지 속성을 만들어본다.
complete_clustering = AgglomerativeClustering(n_clusters = 3, linkage = 'complete')   
average_clustering = AgglomerativeClustering(n_clusters = 3, linkage = 'average') 
```

```python
single_clustering.fit(data)
complete_clustering.fit(data)
average_clustering.fit(data)
```

```python
single_cluster = single_clustering.labels_
complete_cluster = complete_clustering.labels_
average_cluster = average_clustering.labels_
```

```python
print(single_cluster)
print(complete_cluster)
print(average_cluster)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = single_cluster) # single_cluster의 라벨링 값으로 어떻게 그려지는지 본다.
```

```python
plt.scatter(data[:, 0], data[:, 1], c = complete_cluster)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = average_cluster) # <- 아래 있는 실제 데이터의 label과 비교했을 때, 가장 잘 표현했다.
```

* color가 조금씩 바뀌는 것은 상관없음, 얼마나 잘 구별해주는가가 좋다.

```python
plt.scatter(data[:, 0], data[:, 1], c = label)  # <- color에 실제 데이터의 label을 넣어 주었다.
```

```python
# 덴드로그램을 보기 위해 시각화를 해주었다.

from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize = (10, 10))
children = single_clustering.children_  # childre_은 하위 노드에 대한 것이 들어있다.
distance = np.arange(children.shape[0])  # <- 전체의 평균값을 구한 다음에 값을 잡아 넣는다.
no_of_observations = np.arange(2, children.shape[0] + 2)
linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

dendrogram(linkage_matrix, p = len(data), labels = single_cluster,
           show_contracted = True, no_labels = True)
```

```python
# Silhouette

from sklearn.metrics import silhouette_score

best_n = 1 
best_score = -1         # 1이 제일 좋은 값이니깐 -1부터 시작한다.
 
for n_cluster in range(2, 11):
  kmeans = KMeans(n_clusters = n_cluster) # n_cluster가 반복적으로 들어간다.
  kmeans.fit(data) # 비지도 학습이라서 data만 준다. label <- X.
  cluster = kmeans.predict(data) # 예상치를 뽑는다.
  score = silhouette_score(data, cluster) # data하고 cluster를 이용한다.
  print('클러스터의 수: {}, 실루엣 점수: {:.2f}'.format(n_cluster, score))

  if best_score < score:
    best_n = n_cluster
    best_score = score

print('가장 높은 실루엣 점수를 가진 클러스터의 수: {}, 실루엣 점수: {:.2f}'.format(best_n, best_score))
```

```python
# Silhouette (Average)

from sklearn.metrics import silhouette_score

best_n = 1 
best_score = -1         
 
for n_cluster in range(2, 11):
  average_clustering = AgglomerativeClustering(n_clusters = n_cluster, linkage = 'average') 
  average_clustering.fit(data) 
  cluster = average_clustering.labels_  # average_clustering은 predict가 없어서 albels_를 가지고 오면 된다.
  score = silhouette_score(data, cluster) 
  print('클러스터의 수: {}, 실루엣 점수: {:.2f}'.format(n_cluster, score))

  if best_score < score:
    best_n = n_cluster
    best_score = score

print('가장 높은 실루엣 점수를 가진 클러스터의 수: {}, 실루엣 점수: {:.2f}'.format(best_n, best_score))
```

* 클러스터의 수가 3개가 실루엣 점수가 가장 높으니깐 클러스터가 3개라는 것을 알 수 있다.
